## Gaze Estimation introduction
Gaze estimation is performed by capturing images of a user's eye as shown in [Figure 2](#model) and mapping them to screen positions using a function $f_\mathbf{w}(x, y)$. Humans look at things by orienting their eyes so that the light from the point of focus hits the _Fovea_ (a point on the retina). The Fovea is not directly aligned with the center of the pupil, but at a person-specific angle, as shown in [Figure 1](#kappa). The pupil position can be used to infer gaze, but to obtain accurate gaze estimates requires training data (called calibration).

The true nature of the function that transforms pupil positions to screen positions is not easy to model. Even when the head is stationary, the relationship between pupil position in the image and gaze is non-linear due to the pupil's movement along a spherical curve. For now, we'll approximate the gaze mapping by using a linear function. 
Since the goal is to predict a $2D$ screen coordinate from a $2D$ pupil coordinate, it's more straightforward to train two separate models for each coordinate by following the pattern of previous exercises. The linear gaze model can be expressed as:struct

$$
\begin{aligned}
x' &= ax + by + c\\
y' &= dx + ey + f.
\end{aligned}
$$
Gaze mappings is demonstrated in [Figure 3](#gaze) to the right. Here, the $x$ coordinate of the pupil maps to the $x$ coordinate on the screen. In the real model, we use both $x$ and $y$ as inputs to both models estimating the $x$ position and $y$ position on the screen.

***Answer for 1.*** *Identify the model parameters, inputs and outputs for the model.*
\
If  $f_{w_1}(x, y) = x^{\prime} = ax + by + c$, then it $f_{w_1}(x, y)$ can be rewritten to 
$f_{w_1}(x, y) =    \begin{bmatrix}
                        a & b & c
                    \end{bmatrix}
                    \begin{bmatrix}
                        x \\
                        y \\
                        1
                    \end{bmatrix} = x^{\prime}$.

Given an input $A_1$, output $x^{\prime}$ and model parameters $\vec{w_1}$, then $f_{\vec{w_1}}(x, y)$ can be rewritten to:

$$
    \begin{align*}
        x^{\prime} = f_{\vec{w_1}}(x, y) 
        &= A_1 \vec{w_1}
        \\
        &= \begin{bmatrix} x & y & 1 \end{bmatrix} \begin{bmatrix} a \\ b \\ c \end{bmatrix} 
        \\
        &= ax + by + c
    \end{align*}
$$

The same can be done for $f_{\vec{w_2}}(x, y)$:

$$
    \begin{align*}
        y^{\prime} = f_{\vec{w_2}}(x, y)
        &= A_2 \vec{w_2}
        \\
        &= \begin{bmatrix} x & y & 1 \end{bmatrix} \begin{bmatrix} d \\ e \\ f \end{bmatrix}
        \\
        &= dx + ey + f
    \end{align*}
$$


By the definition of model paramters, $\vec{w_1}$ and $\vec{w_2}$ are therefore the model paramters.

***Answer for 2.*** *Identify and determine the minimim number of data points needed to fit the model.*

We can combine the 2 functions $f_{\vec{w_1}}$ and $f_{\vec{w_2}}$ in the following function where $b = \begin{bmatrix} x^{\prime} \\ y^{\prime} \end{bmatrix}$:

$$
    \begin{align*}
        f_{\vec{w}}(x, y) &= b
        \\
        &= A \vec{w}
        \\
        &=  \begin{bmatrix} 
                x & y & 1 & 0 & 0 & 0 \\
                0 & 0 & 0 & x & y & 1
            \end{bmatrix}
            \begin{bmatrix} 
                a \\ b \\ c \\ d \\ e \\ f
            \end{bmatrix}
        \\
        &=  \begin{bmatrix}
                ax + by + c \\
                dx + ey + f
            \end{bmatrix}
    \end{align*}
$$

Mathematically, to isolate $\vec{w}$ through $\mathbf{w} = A^{-1}b$, where $A$ contains input values and $b$ contains output values:

$$
    \begin{align}
        \vec{w} &= A^{-1}b
        \\
        &= \begin{bmatrix}
            x_1 & y_1 & 1 & 0 & 0 & 0 \\
            0 & 0 & 0 & x_1 & y_1 & 1 \\
            x_2 & y_2 & 1 & 0 & 0 & 0 \\
            0 & 0 & 0 & x_2 & y_2 & 1 \\
            x_3 & y_3 & 1 & 0 & 0 & 0 \\
            0 & 0 & 0 & x_3 & y_3 & 1
        \end{bmatrix}^{-1}
        \begin{bmatrix}
            x_1^{\prime} \\
            y_1^{\prime} \\
            x_2^{\prime} \\
            y_2^{\prime} \\
            x_3^{\prime} \\
            y_3^{\prime}
        \end{bmatrix}
    \end{align}
$$

Therfore, since the we need 3 pairs of $x$-$y$ data points to create a square matrix of $f_{\vec{w}}$, then we only need a minimum of 3 data points to train our models. This makes sense, since 3 is also the amount of paramters we have ($x$, $y$, and $1$).

***Answer for 3.*** *Using the least amount of points required to fit the model, does it matter which points are used? Why? Why not?*
\
We need points which can be used to isolate the unknowns in $\vec{w_1}$ and $\vec{w_2}$.
For this, the points used need to be linearly independant. This means that there cannot exist a nontrivial (at least one of the coefficients is nonzero) linear combination of the points as vectors that results in the zero vector.

Furthermore, the model will be tailored specifically to the points used. If they are very close to each other, the model will likely fit poorly, because the noise of the points will have a greater effect.

```python
dataset_name = {0: "grid ", 1: "circle ", 2: "random ", 3: "line "}

def get_design_matrix(pupil_positions : np.ndarray[np.float32, np.float32]):
    """
    Constructs a design matrix.
    Parameters:
    pupil_positions (N x 2 numpy array): n samples, 2 features (px, py).

    Returns:
    design_matrix (N x 3 numpy array) : n_samples, 2 features + 1
    """

    # Write your implementation

    one_vals = np.ones(len(pupil_positions))
    A = np.c_[one_vals, pupil_positions] # numpy.c_ = Translates slice objects to concatenation along the second axis.
    return A



def linear(param, x):
    return param[0] + param[1]*x[0] + param[2]*x[1]

class GazeModel:
    def calibrate(self, dataset):
        ...

    def predict(self, pupil):
        ...

    def predict_many(self, pupils):
        return np.array([self.predict(pupil) for pupil in pupils])


class LinearGaze(GazeModel):
    def calibrate(self, dataset : dict, pupil_training_set : str, screen_training_set : str):
        """
            Calibrates the LinearGaze model by fitting two linear regression models for x and y coordinates based
            on the training dataset.

            Parameters:
                dataset (dict): The dictionary D containing the training data with two keys:
                    -  "pupils_train " (N x 2 numpy array): n samples, 2 features (px, py) representing pupil
                    positions.
                    -  "positions_train " (N x 2 numpy array): n samples, 2 features (screen coordinate x, screen coordinates y) representing
                    the true gaze positions being the screen coordinates.

            Returns:
                None: Updates the model's parameters (self.model_x, self.model_y) based on
                the least squares solution for x and y positions.
        """
        
        # Write your implementation
        pupil_set : np.ndarray = dataset[pupil_training_set]
        screen_set : np.ndarray = dataset[screen_training_set]

        # Design matrix for pupil training set
        A = get_design_matrix(pupil_set)
        x, residuals, rank, s = np.linalg.lstsq(A, screen_set[:,0],rcond=None)
        y, residuals, rank, s = np.linalg.lstsq(A, screen_set[:,1],rcond=None)

        self.model_x = x
        self.model_y = y


    def predict(self, pupil : np.ndarray):
        """
        Predicts the gaze position (x, y) based on the given pupil position using the calibrated model.

        Parameters:
            pupil (1 x 2 numpy array): 2 features (px, py) representing the current pupil position.

        Returns:
            x, y (float, float): Predicted gaze position (x, y) based on the linear model.
        """

        # Write your implementation
        w_1 = self.model_x
        w_2 = self.model_y
        predicted_x = linear(w_1, pupil)
        predicted_y = linear(w_2, pupil)
        return (predicted_x, predicted_y)

# for i in range(4):
#     dataset = D[i]
#     lg = LinearGaze()
#     lg.calibrate(dataset,  "pupils_n_train ",  "screen_coordinates_n_train ")
#     pupil = dataset[ "pupils_n_test "]

#     screen = dataset[ "screen_coordinates_n_test "]
#     print(pupil[0])
#     print(screen[0])
#     print(lg.predict(pupil[0]))
#     print()
```

```python
def results_for_model(model, dataset_train, dataset_test, pupil_training_set, screen_training_set, pupil_test_set, screen_test_set, d_max):
    """
    Evaluates the performance of the model across multiple datasets and computes prediction errors.

    Parameters:
        model (object): A gaze model object.
        d_max (int, optional): Maximum number of patterns of a dataset to evaluate. Default is 4.

    Returns:
        results (dict): A dictionary where each key corresponds to a dataset index (0 to d_max-1), and each value is a dictionary containing:
            - "errors" (N x 2 numpy array): Absolute errors between predicted and ground truth positions (x, y).
            - "rmse" (float): Root Mean Squared Error (RMSE) for the predictions across the dataset.
            - "dist" (float): Mean Euclidean distance between the predicted and actual positions.
            - "mae" (1 x 2 numpy array): Mean absolute error in x and y dimensions.
    """


    results = {}
    for d in range(d_max):
        train = dataset_train[d]
        test = dataset_test[d]
        model.calibrate(train, pupil_training_set, screen_training_set)
        predicted = model.predict_many(test[pupil_test_set])
        ground_truth = test[screen_test_set]

        dist_array = np.array([np.linalg.norm(x) for x in predicted-ground_truth])
    
        res = {}
        res["predicted"] = predicted
        res["ground_truth"] = ground_truth
        res["model_x"] = model.model_x
        res["model_y"] = model.model_y
        res["errors"] = abs(predicted-ground_truth)
        res["rmse"] = np.sqrt(np.mean((predicted-ground_truth)**2))
        res["dist"] = np.mean(dist_array)
        res["mae"] =  [np.mean(abs(predicted[:,0] - ground_truth[:,0])), np.mean(abs(predicted[:,1] - ground_truth[:,1]))]
        results[d] = res
    return results
    

 
results_linear = results_for_model(LinearGaze(), D, D,  'pupils_n_train', 'screen_coordinates_train', 'pupils_n_test', 'screen_coordinates_test', 4)
```

$a$ and $b$ are connected in the sense that if $a$ increases, but $b$ stays the same, the error will increase. 
\
For $b$, if it is decreases, where $a$ stays the same, the error will also increase. 
\
So, to keep the error low, $a$ needs to increase AND $b$ needs to decrease.

In other words, there is a relation between $a$ and $b$ where the error value is as low as possible when $a$ and $b$ follow this relation.

For $c$, we can see that it is $a$ parabola, which means it has $a$ distinct least value, which would be the optimal one, if $a$ and $b$ stays the same.
